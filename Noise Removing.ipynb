```python
import skimage.io as io
import cv2 as cv
import numpy as np 
from enum import Enum
from skimage.filters import gaussian
from skimage.filters import median
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
from skimage.util import random_noise
```

```python
# ////////////// <========> Show Images Function <========> ////////////// #

def show_images(images,titles=None):
    #This function is used to show image(s) with titles by sending an array of images and an array of associated titles.
    # images[0] will be drawn with the title titles[0] if exists
    # You aren't required to understand this function, use it as-is.
    n_ims = len(images)
    if titles is None: titles = ['(%d)' % i for i in range(1,n_ims + 1)]
    fig = plt.figure()
    n = 1
    for image,title in zip(images,titles):
        a = fig.add_subplot(1,n_ims,n)
        if image.ndim == 2: 
            plt.gray()
        plt.imshow(image)
        a.set_title(title)
        n += 1
    fig.set_size_inches(np.array(fig.get_size_inches()) * n_ims)
    plt.show()
    
# //////// <====> End <====> //////// #
```

```python
pyramids = io.imread('pyramids.jpeg')
gray=cv.cvtColor(pyramids,cv.COLOR_BGR2GRAY)

noise = random_noise(gray, mode='gaussian')

img1 = median(noise)
img2 = gaussian(noise, sigma=3.5, output=None, mode='nearest')
show_images([pyramids,gray,noise,img1,img2])
```

```python
# ////////////// <========> Gaussian Filter <========> ////////////// #
def gaussian_filter(img):
    filter = np.array([[1,2,1],[2,4,2],[1,2,1]])
    return convolve2d(img,filter)
# //////// <====> End <====> //////// #
```
